#### Выполнить прямое и обратное преобразование Фурье для заданных отсчетов используя готовые библиотеки в Python

```python
#базовые библиотеки
import numpy as np
import cmath
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
```

```python
u=[2,12,60,50] #начальная функция
u_fft=fft(u) #БПФ
u_ifft=ifft(u_fft)# обратное БПФ
u_fft
```

```python
cmath.phase(u_fft[3])+np.pi
```

```python
#строим
plt.figure(figsize=(15, 3))
k=3
plt.subplot(131)
plt.plot(u,'o--')
plt.title("u")

plt.subplot(132)
plt.plot(abs(u_fft),'o--')
plt.title("Модуль БПФ")

plt.subplot(133)
plt.plot(abs(u_ifft),'o--')
plt.title("Обратное БПФ")
plt.show()
```

```python
print('функция:',u)
print('БПФ:',u_fft)
print('Обратное БПФ:',u_ifft)
```
